```python
train = []
y_train = []
for i in range(n):
    train.append(word_tokenize(df["text"][i].lower()))
    y_train.append(df["airline_sentiment"][i])
```

```python
test = []
y_test = []
for i in range(n1):
    test.append(word_tokenize(df["text"][i].lower()))
    y_test.append(df["airline_sentiment"][i])
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
train
```

```python
train_documents = [clean_review(document) for document in train]
```

```python
test_documents = [clean_review(document) for document in test]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text_documents = [" ".join(document) for document in train_documents]
test_text_documents = [" ".join(document) for document in test_documents]
```

```python
test_text_documents
```

```python

#x_train, x_test, y_train, y_test = train_test_split(text_documents, y_train)
```

```python
count_vec = CountVectorizer(max_features = 2000,max_df = 0.9,ngram_range=(1,2))
x_train_features = count_vec.fit_transform(text_documents)
```

```python
x_test_features = count_vec.transform(test_text_documents)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(random_state=0)
clf.fit(x_train_features, y_train)
test_predict = clf.predict(x_test_features)
```
